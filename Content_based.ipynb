```python
import numpy as np
import pandas as pd
import nltk
nltk.download(["punkt","stopwords","wordnet"])
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
genres = mov["genres"]
li=[]
for i in range(len(genres)):
    temp = genres[i].lower()
    temp = temp.split("|")
    temp = [lemmatizer.lemmatize(word) for word in temp]
    li.append(" ".join(temp))
```

```python
print(li)
movies_dataset = pd.DataFrame(li,columns=["genres"],index=mov["title"])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(mov["genres"]).toarray()
X
```

```python

output = mov.loc[:,['movieId','title']]
output = output.join(pd.DataFrame(X))
output
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(X) 
similarities
```

```python

uid = 1 #For user 18 lets recommend movies based on his recent watched movie
time = rat.loc[rat["userId"]==uid,["movieId","timestamp"]]
latest_movieId_watched_by_user = time.sort_values(by="timestamp",ascending=False)["movieId"].values[0]

latest_movieId_watched_by_user
```

```python

movie_index = mov.loc[mov['movieId']==latest_movieId_watched_by_user,["title"]].index[0]
output.loc[output['movieId']==161,:]
```

```python
movie_index = mov.loc[mov['movieId']==latest_movieId_watched_by_user,["title"]].index[0]
similarity_values = pd.Series(similarities[movie_index])

similarity_values.sort_values(ascending=False)
```

```python
similar_movie_indexes = list(similarity_values.sort_values(ascending=False).index)
similar_movie_indexes
```

```python
similar_movie_indexes.remove(movie_index)
```
